```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
from ast import literal_eval
import textwrap
import graphviz
import pygraphviz
from pyvis.network import Network
# Read data from TSV file
df = pd.read_csv("biorxiv_cancer_posts (2).tsv", delimiter="\t")

df_1 = df.drop_duplicates(subset=["Title"])
df_1 = df_1.sort_values('Collection Period', ascending = False)
df_sampled = df_1.iloc[:25,:]
#df_sampled = df_1.sample(n=20, random_state=42)

G = nx.Graph()
attributes = defaultdict(list)

for index, row in df_sampled.iterrows():
    title = row['Title']
    keywords = str(row['Keywords']).split(" ")
    for keyword in keywords:
        attributes[keyword].append(title)

attributes = {key: list(set(value)) for key, value in attributes.items()}
attributes = {key: value for key, value in attributes.items() if len(value) >= 2}

for keyword in attributes.keys():
    G.add_node(keyword, type='keyword')

for title in set(element for sublist in attributes.values() for element in sublist):
    wrapped_title = textwrap.fill(title, width=20)  # Adjust width as needed
    G.add_node(wrapped_title, type='title')

for keyword, related_titles in attributes.items():
    for title in related_titles:
        wrapped_title = textwrap.fill(title, width=20)  # Adjust width as needed
        G.add_edge(wrapped_title, keyword)

# Define node colors and sizes based on type
node_color = {'keyword': 'lightgreen', 'title': 'skyblue'}
node_size = {'keyword': 10000, 'title': 15000}

# Assign node attributes for color and size
for node, data in G.nodes(data=True):
    node_type = data['type']
    G.nodes[node]['color'] = node_color[node_type]
    G.nodes[node]['size'] = node_size[node_type]

# Draw the graph
pos = nx.nx_agraph.graphviz_layout(G, "sfdp")
#plt.figure(figsize=(40, 40))

# Draw nodes
#for node, data in G.nodes(data=True):
#    nx.draw_networkx_nodes(G, pos, nodelist=[node], node_color=data['color'], node_size=data['size'])

# Draw edges
#nx.draw_networkx_edges(G, pos, edge_color='gray')

# Draw labels
#nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

#plt.title("Graph of Paper Titles and Common Keywords")
#plt.axis('off')  # Turn off axis
#plt.show()
node_size = {'keyword': 10, 'title': 15}

# Assign node attributes for color and size
for node, data in G.nodes(data=True):
    node_type = data['type']
    G.nodes[node]['color'] = node_color[node_type]
    G.nodes[node]['size'] = node_size[node_type]

g = Network(height = 800, width = 1500, notebook = True)

g.barnes_hut()
for node, data in G.nodes(data=True):
    if data['type'] == 'title':
        g.add_node(node, size=60, color='skyblue', font={'size': 40})
    elif data['type'] == 'keyword':
        g.add_node(node, size=80, color='lightgreen', font={'size': 100})
for edge in G.edges():
    g.add_edge(edge[0], edge[1], width=12)
#g.from_nx(G)
g.show("ex.html")
```

```python
g.set_options("""
var options = {
  "nodes": {
    "font": {
      "size": 4
    },
    "scaling": {
      "min": 10,
      "max": 20
    }
  }
}
""")

# Initialize a new graph
G = nx.Graph()

# Define datasets and attributes
datasets = [
    "1000 Genomes Project",
    #"American Gut Project",
    #"Jump Cell Painting",
    "Cancer Cell Line Encyclopedia (CCLE)",
    #"CytoImageNet",
    "ENCODE Project",
    "Gene Ontology Annotations",
    "The Cancer Genome Atlas Program"
    #"RCSB Protein Data Bank"
]

attributes = {
    "Next-Generation Sequencing": ["1000 Genomes Project", "ENCODE Project", "The Cancer Genome Atlas Program"],
    "Human Genome": ["1000 Genomes Project", "ENCODE Project"],
    #"Microbiome": ["American Gut Project"],
    #"Cell Painting Assay": ["Jump Cell Painting"],
    #"High-Content Imaging": ["Jump Cell Painting", "CytoImageNet"],
    "Cancer": ["Cancer Cell Line Encyclopedia (CCLE)", "The Cancer Genome Atlas Program"],
    "EGFR":["Cancer Cell Line Encyclopedia (CCLE)", "Gene Ontology Annotations", "The Cancer Genome Atlas Program"]
    #"Functional Genomics": ["ENCODE Project", "Gene Ontology Annotations"],
    #"Protein Structures": ["RCSB Protein Data Bank"],
    #"Various Genes": ["Gene Ontology Annotations"]
}

# Add nodes for datasets and attributes
for dataset in datasets:
    G.add_node(dataset, type='dataset')

for attribute in attributes.keys():
    G.add_node(attribute, type='attribute')

# Add edges between datasets and attributes
for attribute, related_datasets in attributes.items():
    for dataset in related_datasets:
        G.add_edge(dataset, attribute)

# Draw the graph
pos = nx.kamada_kawai_layout(G)
plt.figure(figsize=(24, 24))

# Node color and size based on type
node_color = ['skyblue' if G.nodes[node]['type'] == 'dataset' else 'lightgreen' for node in G]
node_size = [1800 if G.nodes[node]['type'] == 'dataset' else 1500 for node in G]

nx.draw(G, pos, with_labels=True, node_color=node_color, node_size=node_size, edge_color='gray', font_size=10, font_weight='bold')
plt.title("Graph of Datasets and Common Attributes")
plt.show()
```
