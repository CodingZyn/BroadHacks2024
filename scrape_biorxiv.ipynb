```python
import requests
import json
import pandas as pd

endpoint_format = "https://api.biorxiv.org/details/[server]/[interval]/[cursor]/[format]"
server = "biorxiv"
format = "json"
category = "cancer biology"
num_requested = 100
num_in_category = 0
df_list = []
cursor = 0
start_date = "2024-06-20"
end_date = "2024-06-27"
while True:
    interval = f"{start_date}/{end_date}"
    endpoint = endpoint_format.replace("[server]", server).replace("[interval]", interval).replace("[cursor]", str(cursor)).replace("[format]", format)

    response = requests.get(endpoint)
    data = response.json()
    if data["collection"] and num_in_category < num_requested:
        num_found_this_cursor = 0
        for doc in data["collection"]:
            if category in doc["category"] and doc not in df_list:
                num_in_category += 1
                num_found_this_cursor += 1
                df_list.append(doc)
        cursor += 1
        print(f"Found {num_in_category} papers in {category=} so far")
        if num_found_this_cursor == 0:
            print("No more papers in this interval")
            # move back start date
            end_date = start_date
            start_date = pd.to_datetime(start_date) - pd.DateOffset(7)
            start_date = start_date.strftime("%Y-%m-%d")
            end_date = pd.to_datetime(end_date) - pd.DateOffset(7)
            end_date = end_date.strftime("%Y-%m-%d")
            cursor = 0
    else:
        break

print(f"Number of papers in {category}: {num_in_category}")
```

```python
df = pd.DataFrame(df_list)
df
```

```python
# attributes we need: title, description, keywords, dataset type, collection period, organism, genes, tissue/cell type, condition, technique, instrument platform, software, usage restrictions, related datasets
# use LLM to extract these attributes from the title, abstract, and keywords

from keybert import KeyBERT

model = KeyBERT('distilbert-base-nli-mean-tokens')

def extract_keywords(text):
    keywords = model.extract_keywords(text)
    # remove confidence
    keywords = [keyword[0] for keyword in keywords]
    return keywords

# extract attributes from the title, abstract, and keywords
df["title_attributes"] = df["title"].apply(extract_keywords)
df["abstract_attributes"] = df["abstract"].apply(extract_keywords)
df
```

```python
# print duplicates
duplicates = df[df.duplicated(subset=["title", "abstract"], keep=False)]
duplicates
```

```python
df_dropped = df.copy()
df_dropped["Title"] = df["title"]
df_dropped["Description"] = df["abstract"].replace(r"\n", " ", regex=True)
df_dropped["Keywords"] = (df["title_attributes"] + df["abstract_attributes"] + df["category"].apply(lambda x: [x])).apply(lambda x: " ".join(list(set(word.replace(" ", "_") for word in x))))
df_dropped["Dataset Type"] = ""
df_dropped["Collection Period"] = df["date"]
df_dropped["Organism"] = ""
df_dropped["Genes"] = ""
df_dropped["Tissue/Cell Type"] = ""
df_dropped["Condition"] = ""
df_dropped["Technique"] = ""
df_dropped["Instrument Platform"] = ""
df_dropped["Software"] = ""
df_dropped["Usage Restrictions"] = df["license"]
df_dropped["Related Datasets"] = ""
df_dropped = df_dropped.drop(columns=["title", "abstract", "title_attributes", "abstract_attributes", "doi", "authors", "author_corresponding", "author_corresponding_institution", "date", "version", "type", "server", "published", "jatsxml", "category", "license"])
df_dropped
```

```python
df_dropped.to_csv("biorxiv_cancer.csv", index=False)
```
